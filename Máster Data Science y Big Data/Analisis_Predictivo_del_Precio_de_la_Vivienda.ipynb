# Entrenamiento

***En la hioptesis 1 trabajamos solo con las variables cuantitativas.***

***Para ello eliminamos las cualitativas.***

```python
#Vamos a eliminar las variables cualitativas
categoricas = ['date', 'street', 'city', 'statezip', 'country']
df_hipot1 = df.drop(categoricas, axis = 1)
```

```python
#Lo comprobampos
df_hipot1.head()
```

***Para poder hacer el entrenamiento tenemos que definir nuestra variable dependiente que será el precio (price). La eliminamos del eje X y la añadimos en el eje Y.***

```python
x_hipot1 = df_hipot1.drop('price',axis= 1)#variables independientes
y_hipot1 = df_hipot1['price']# variable dependiente
```

```python
# Preparamos el train data y el test data
from sklearn.model_selection import train_test_split
x_hipot1_train, x_hipot1_test, y_hipot1_train, y_hipot1_test = train_test_split(x_hipot1, y_hipot1, test_size=0.20, random_state=43)
```

```python
#Entrenamiento
from sklearn.linear_model import LinearRegression
regresion_lineal = LinearRegression()
regresion_lineal.fit(x_hipot1_train, y_hipot1_train)
```

```python
# Validación con el error cuadratico medio
from sklearn.metrics import mean_squared_error
```

```python
#predecimos los datos para los valores usados en el entrenamiento
prediccion_entrenamiento = regresion_lineal.predict(x_hipot1_train)

#Calculamos el MSE
mse_hipot1_train = mean_squared_error(y_true = y_hipot1_train, y_pred = prediccion_entrenamiento)
print("El error cuadratico medio de TRAIN es: " + str(mse_hipot1_train))

prediccion_entrenamiento = regresion_lineal.predict(x_hipot1_test)
#Calculamos el MSE para la validación
mse_hipot1_test = mean_squared_error(y_true = y_hipot1_test, y_pred = prediccion_entrenamiento)
print("El error cuadratico medio de TEST es: " + str(mse_hipot1_test))
```

***Entrenamos nuestro modelo.***

***Le validamos con el error cuadrático medio. (MSE) En el resultado del MSE vemos que trabajar solo con las variables numéricas no nos dan buen resultado. Podriamos intentar convertir las variables categoricas en numéricas, pero como se nos darán mas de 4000 columnas, porque puede haber muchos valores, probamos de otra manera.***

***2. Utilizando variables categoricas y convirtiendolas en  variables numéricas. Eliminamos la variable STREET, porque como tenemos el Código postal, la otra no la consideramos relevante. Por la misma razon eliminamos City.***

```python
#Convertir las categoricas en numéricas
categoricas2 = ['date', 'statezip', 'country']
```

```python
#Preparamos el dataset para la Hipotesis 2
df_hipot2 = pd.get_dummies(data = df, prefix= 'OHE', prefix_sep= '_',
columns = categoricas2, drop_first= True, dtype = 'int8')
```

```python
df_hipot2 = df_hipot2.drop(['street', 'city'], axis = 1)
```

```python
#Variables independientes y dependientes
x_hipot2 = df_hipot2.drop('price',axis= 1)#variables independientes
y_hipot2 = df_hipot2['price']# variable dependiente
```

```python
#Lo comprobamos 
df_hipot2.head()
```

```python
# Preparamos el train data y el test data
from sklearn.model_selection import train_test_split
x_hipot2_train, x_hipot2_test, y_hipot2_train, y_hipot2_test = train_test_split(x_hipot2, y_hipot2, test_size=0.20, random_state=43)
```

```python
#Entrenamiento
from sklearn.linear_model import LinearRegression
regresion_lineal = LinearRegression()
regresion_lineal.fit(x_hipot2_train, y_hipot2_train)
```

```python
#predecimos los datos para los valores usados en el entrenamiento
prediccion_entrenamiento = regresion_lineal.predict(x_hipot2_train)

#Calculamos el MSE
mse_hipot2_train = mean_squared_error(y_true = y_hipot2_train, y_pred = prediccion_entrenamiento)
print("El error cuadratico medio de TRAIN es: " + str(mse_hipot1_train))
print("El error cuadratico medio de TRAIN es: " + str(mse_hipot2_train))
print()
prediccion_entrenamiento = regresion_lineal.predict(x_hipot2_test)
#Calculamos el MSE para la validación
mse_hipot2_test = mean_squared_error(y_true = y_hipot2_test, y_pred = prediccion_entrenamiento)
print("El error cuadratico medio de TEST es: " + str(mse_hipot1_test))
print("El error cuadratico medio de TEST es: " + str(mse_hipot2_test))
```

***Hemos planteado una segunda hipotesis: Trabajar con todas las variables cualitativas y cuantitativas. Para ello las cualitativas las tenemos que transformar en numéricas. Lo haremos con la tecnica one_hot_encoding. Al comprobarlo vemos que se nos han creado 158 columnas, cada una lleva o 0-s o 1-s. El dato correcto es el que lleva 1-s y que todas las variables son numéricas. Hemos eliminado las variables 'street' y city, porque nos parece información redundante.***
